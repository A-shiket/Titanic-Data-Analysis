```python
#Data Analysis and Visualization of the Titanic Dataset Using Seaborn#
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  Data Exploration--
```

```python
titanic=sns.load_dataset('titanic') 
```

```python
titanic
```

```python
#exploring the data with top 10 rows 
titanic.head(10) 
```

```python
#getting information about the data , its data types & columns
titanic.info() 
```

```python
#checking the number of null values under each column in the dataset
titanic.isnull().sum() 
```

```python
##  Data Cleaning ---------
```

```python
#deck column is having mostly null value thats why removing it tends to be the best approach 
titanic=titanic.drop(columns=['deck'])
```

```python
titanic
```

```python
 #replacing null values in age by the mean of all ages
titanic['age']=titanic['age'].fillna(titanic['age'].mean()) #replacing null values in age by the mean of all ages
```

```python
titanic
```

```python
#handeling all the null values
titanic.isnull().sum() 
```

```python
 #viewing the specific rows that contains the remaining null values 
titanic[titanic['embarked'].isnull()]
```

```python
#dropping 2 rows that are containing null values as it has negligible contibrution compared to all data set 
titanic = titanic.dropna(subset=['embarked', 'embark_town']) 
```

```python
titanic
```

```python
# One-hot encode the 'sex' and 'embarked' columns for converting categorical data to numerical value so that it can be used for correlation analysis. This conversion helps in analyzing patterns and relationships across categorical data.
titanic_encoded = pd.get_dummies(titanic, columns=['sex', 'embarked'], drop_first=True)
```

```python
titanic_encoded
```

```python
 #dropping redundant columns which are already present with other names having numerical datatypes which is more useful for data analysis 
titanic_encoded=titanic_encoded.drop(columns=['alive','adult_male','class','alone'])
```

```python
titanic_encoded
```

```python
#renaming certain columns for better understanding & Column Profiling 
titanic_encoded= titanic_encoded.rename(columns={
    'survived': 'Survival_Code',
    'pclass': 'Passenger_Class',
    'sibsp': 'Siblings/Spouses_Aboard',
    'parch': 'Parents/Children_Aboard',
    'who': 'Passenger_Type'
})
```

```python
#Creating a different column for survival status to use it for data visualization 
titanic_encoded['survival_status'] = titanic_encoded['survived'].replace({0: 'Dead', 1: 'Alive'})
```

```python
titanic_encoded
```

```python
###  Data visualization--------
```

```python
sns.scatterplot(x='age', y='fare', data=titanic_encoded, hue='survival_status', palette=['#e74c3c', '#3498db'])
plt.title('Fare vs. Age by Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survival Status')
plt.show()

# A scatter plot was used to examine the relationship between age and fare among survivors and non-survivors. 

# Color-coded by survival status,this visualization provides insights into whether fare and age impacted survival rates. 
# Observation == Younger passengers with higher fares had a greater likelihood of survival.
```

```python
sns.barplot(x='Passenger_Class',y='Survival_Code',data=titanic_encoded)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('')
plt.ylabel('Average Survival rate')
plt.xticks([0,1,2],['First Class','Second Class','Third Class'])
plt.show()

#Analyzing survival rates across passenger classes reveals a clear trend.

#Observation == first-class passengers had a significantly higher survival rate.
#This pattern highlights the influence of socio-economic status on survival, as passengers in higher classes were more likely to survive.
```

```python
sns.barplot(x='sex_male', y='Survival_Code', data=titanic_encoded, hue='sex_male', palette=['#3498db', '#e74c3c'], dodge=False,legend=False)
plt.title('Survival Rate by Gender')
plt.ylabel('Average Survival Rate')
plt.xlabel('')
plt.xticks([0, 1], ['Female', 'Male'])
plt.show()

# Survival rates by gender 

# Observation == It reveals that women were more likely to survive than men, 
# possibly due to 'women and children first' protocols. This bar plot clearly shows the disparity in survival between genders.
```

```python
embark_survival = titanic_encoded.groupby(['embarked_Q', 'embarked_S'])['Survival_Code'].mean()
embark_survival.index = ['Cherbourg', 'Queenstown', 'Southampton']
embark_survival.plot(kind='bar', color=['#3498db', '#e74c3c', '#2ecc71'])
plt.title('Survival Rate by Embarkation Point')
plt.xticks(rotation=0)#to place the x lables correctly for readablility 
plt.xlabel('')
plt.ylabel('Average Survival Rate')
plt.show()

# This bar plot illustrates survival rates based on embarkation points.

# Observations == Passengers boarding from certain locations, 
#such as Cherbourg, had a higher survival rate, potentially reflecting socio-economic or demographic factors.
```

```python
sns.boxplot(x='sex_male', y='age', data=titanic_encoded,hue='survival_status')
plt.title('Age Distribution by Gender & Survival Status')
plt.xticks([0,1],['Female','Male'])
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()

# A box plot displays age distributions by gender and survival status. 

# Observation == Younger passengers, particularly females, show a higher survival rate, 
#supporting the hypothesis that younger individuals and females were prioritized during evacuation.
```

```python
correlation_matrix = titanic_encoded[['Survival_Code', 'Passenger_Class', 'age', 'fare', 'sex_male']].corr()
key_correlations = correlation_matrix['Survival_Code'].abs().sort_values(ascending=False).iloc[1:5]
print(key_correlations)

#The correlation table provides a quantitative overview of relationships between numerical features in the dataset.
# Strong correlation with Males showed that Females were having higher Survival Rate.
```

```python
sns.histplot(data=titanic,x='age', bins=20, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()

# The age distribution histogram 
# Observation == It shows the majority of passengers were between 20-40 years old, with a gradual decrease in frequency for older passengers.
#This distribution helps in understanding the demographics of the Titanic passengers.
```

```python
sns.pairplot(titanic_encoded, vars=['age', 'fare', 'Passenger_Class'], hue='survival_status')
plt.suptitle('Pair Plot of Key Variables by Survival Status', y=1.02)
plt.show()

# A pair plot allows us to examine relationships and distributions among key variables, such as age, fare, passenger class, and survival status. 
# Observation == By using survival status as a hue, we observe distinct patterns, such as higher survival rates for higher fares and younger ages.
```
